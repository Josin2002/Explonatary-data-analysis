# excercise

```python
#1 no of employees in each team and percentage splitting
team_members=df['Team'].value_counts()
percentage_split=team_members/len(df)*100
```

```python
percentage_split
```

```python
#visulaisation of percentage split
percentage_split.plot(kind='pie')
plt.title('percentage split by teams')
plt.show()
```

```python
#2 to segregate the employees w.r.t different positions
position_counts = df['Position'].value_counts()
```

```python
position_counts
```

```python
#3 to find the age group with the most employees
age_groups=pd.cut(df['Age'],
bins=[18,22,26,30,34,38,42,46,50],
labels=['18-22','22-26','26-30','30-34','34-38','38-42','42-46','46-500'])
most_age_group=age_groups.value_counts().idxmax()
```

```python
most_age_group
```

```python
#4 to identify the team and position with the highest salary spending
high_salary_team_position=df.groupby(['Team','Position'])['Salary'].sum().idxmax()
```

```python
high_salary_team_position
```

```python
#5 visual representation of correlation between age and salary
plt.figure(figsize=(8,6))
sns.scatterplot(x='Age',y='Salary',data=df)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
correlation=df[['Age','Salary']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation,annot=True,cmap='coolwarm',fmt=".3f")
plt.title('Age vs Salary')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Age',y='Salary',data=df)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
